### 1. Line Plot
- plt.plot()

```python
x_axis = [10,12,14,16]
y_axis = [5,6,7,8]
plt.plot(x_axis, y_axis)
```

```python
x_axis = [10,12,14,16]
y_axis = [5,6,7,8]
plt.plot(x_axis, y_axis, label = 'Line')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('Line Graph')
plt.legend(loc = 'upper right')

```

```python
help(plt.legend)
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,20)
x
```

```python
plt.plot(x, x**2, color = 'red', label = 'Line1')
plt.plot(x, x**3, color = 'yellow', label = 'Line2')
plt.plot(x, x**4, color = 'green', label = 'Line3')
# plt.show()
plt.legend()
```

```python
help(plt.plot)
```

```python
x1 = np.arange(100)
x1
```

```python
plt.plot(x1, x1 + 10, color='r', linestyle = '--')
plt.plot(x1, x1 + 15, color='g', linestyle = ':')
plt.plot(x1, x1 + 20, color='m', linestyle = '-.')
plt.plot(x1, x1 + 25, color='b', linestyle = '-')
```

```python
x2 = np.linspace(10,30,50)
x2
```

```python
plt.plot(x2, np.sin(x2), color = 'r', marker = 'o', markersize = 10, markerfacecolor = 'w')
plt.plot(x2, np.cos(x2), color = 'b', marker = 'D', markersize = 10, markerfacecolor = 'gray')
```

```python
plt.plot(x2, np.sin(x2 - 1))
plt.plot(x2, np.sin(x2 - 2))
plt.plot(x2, np.sin(x2 - 3))
plt.plot(x2, np.sin(x2 - 4))
plt.plot(x2, np.sin(x2 - 5))
plt.plot(x2, np.sin(x2 - 6))
```

### 2. Bar Chart
- plt.bar()

```python
Year = ['2019 Dec','2020 Jan','2020 March','2020 July']
GoldPrice = [1000,3000,5000,5300]
plt.bar(Year, GoldPrice, color = 'r')
# Apply title, labes
```

```python
Year = ['2019 Dec','2020 Jan','2020 March','2020 July']
Gold = [1000,3000,5000,5300]
silver = [500,300,400,1000]
plt.bar(Year,Gold,width = 0.33)
plt.bar(Year,silver,width = 0.33)
```

```python
x = np.arange(5)
x
```

```python
plt.bar(x, [90,82,55,67,77], width = 0.33)
plt.bar(x + 0.33, [23,90,92,44,30], width = 0.33)
```

#### Horizontal Bar Graph
- plt.barh()

```python
plt.barh(x, [90,82,55,67,77], color = 'r')
```

```python
x = [1,2,3,4,5]
y = [10,55,23,44,67]
plt.bar(x,y, color = ['r','gray','green','m','pink'], edgecolor = 'blue')
```
